```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib import rcParams
```

```python
rcParams['font.family']='Times New Roman'
```

```python
valid = pd.read_csv('./spss/valid.csv',sep='\t', header=None).dropna().values
back = pd.read_csv('./spss/back_color.csv',sep='\t', header=None).dropna().values
```

```python
fig = plt.figure(figsize=(12,5))
plt.subplot(121)
plt.hist(back[:,1], range = (0,1), bins=40, density=True, alpha = 0.8, label='Non-Professional Traders')
plt.hist(back[:,0], range = (0,1), bins=40, density=True, alpha = 0.8, label='Professional Traders')
plt.xlabel('Ratio of Fixations on Red Candlesticks')
plt.ylabel('Percentage')
plt.legend()

plt.subplot(122)
plt.hist(valid[:,1], range = (0,1), bins=40, density=True, alpha = 0.8,label='Non-Professional Traders')
plt.hist(valid[:,0], range = (0,1), bins=40, density=True, alpha = 0.8,label='Professional Traders')
plt.xlabel('Valid Ratio of Fixations')
plt.legend()
plt.tight_layout()

fig.savefig('content.svg')
```
